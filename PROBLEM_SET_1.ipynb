### Problem 1

Suppose that the data for analysis includes the attribute **age**. The age values for the data tuples are (in increasing order):

$$13, 15, 16, 16, 19, 20, 20, 21, 22, 22, 25, 25, 25, 25, 30, 33, 33, 35, 35, 35, 35, 36, 40, 45, 46, 52, 70$$

The dataset consists of **$n = 27$** age values.

Answer:

 (a) Mean and Median

**Mean:**

$$\text{Mean} = \frac{\sum \text{age}}{n} = \frac{809}{27} \approx 29.96$$

**Median:**

$$\text{Median} = \text{14th value} = \mathbf{25}$$



(b) Mode and Modality

- **Mode:** The values that appear most frequently. In this dataset, both **25** and **35** each appear **4 times**.
- **Modality:** Since there are two modes, the data is **bimodal**.


(c) Midrange



$$\text{Midrange} = \frac{\text{min} + \text{max}}{2} = \frac{13 + 70}{2} = \mathbf{41.5}$$



(d) First Quartile ($Q_1$) and Third Quartile ($Q_3$)



- **Lower Half** (values 1 to 13): The median is the **7th value**.
$$Q_1 = \text{7th value} = \mathbf{20}$$

- **Upper Half** (values 15 to 27): The median is the **21st value**.
$$Q_3 = \text{21st value} = \mathbf{35}$$

**Interquartile Range (IQR):**

$$\text{IQR} = Q_3 - Q_1 = 35 - 20 = 15$$



(e) Five-Number Summary

The five-number summary provides a quick overview of the data distribution:

| Statistic | Value |
|-----------|-------|
| Minimum | 13 |
| First Quartile ($Q_1$) | 20 |
| Median ($Q_2$) | 25 |
| Third Quartile ($Q_3$) | 35 |
| Maximum | 70 |



(f) Quantile Plot vs. Quantile-Quantile (Q-Q) Plot

Quantile Plot

- Displays the distribution of a **single dataset**.
- Plots each data value against its corresponding **quantile fraction** (ranging from 0 to 1).
- Helps visualize the **spread**, **skewness**, and **median** of one variable.
- Answers the question: *"How is this dataset distributed?"*
Quantile-Quantile (Q-Q) Plot

- Used to **compare two distributions**.
- Plots the quantiles of one dataset against the quantiles of another dataset (or a theoretical distribution, e.g., the Normal distribution).
- If the plotted points fall **along a straight diagonal line**, the two distributions are similar.
- Answers the question: *"Do these two distributions match?"*


| Feature | Quantile Plot | Q-Q Plot |
|---------|--------------|----------|
| Datasets used | One | Two |
| Purpose | Visualize single distribution | Compare two distributions |
| X-axis | Quantile fraction (0–1) | Theoretical quantiles |
| Y-axis | Observed data values | Observed data values |
| Straight line means | — | Distributions are similar |

### Problem 4

Given two objects represented by the tuples (22, 1, 42, 10) and (20, 0, 36, 8):

(a) Compute the Euclidean distance between the two objects.  

(b) Compute the Manhattan distance between the two objects.  

(c) Compute the Minkowski distance between the two objects, using $h = 3$.  

(d) Compute the supremum distance between the two objects.


###Answer:

two objects represented by tuples  
$A = (22, 1, 42, 10)$ and $B = (20, 0, 36, 8)$

$$
|22 - 20| = 2
$$
$$
|1 - 0| = 1
$$
$$
|42 - 36| = 6
$$
$$
|10 - 8| = 2
$$



(a) Euclidean Distance



$$
d(A, B) = \sqrt{(2)^2 + (1)^2 + (6)^2 + (2)^2}
$$
$$
d(A, B) = \sqrt{4 + 1 + 36 + 4} = \sqrt{45} \approx \mathbf{6.71}
$$



(b) Manhattan Distance



$$
d(A, B) = |2| + |1| + |6| + |2| = \mathbf{11}
$$



(c) Minkowski Distance ($h = 3$)



$$
d(A, B) = (|2|^3 + |1|^3 + |6|^3 + |2|^3)^{1/3}
$$
$$
d(A, B) = (8 + 1 + 216 + 8)^{1/3} = (233)^{1/3} \approx \mathbf{6.15}
$$



(d) Supremum Distance



$$
d(A, B) = \max(|2|, |1|, |6|, |2|) = \mathbf{6}
$$

# Problem 5

It is important to define or select similarity measures in data analysis. However, there is no commonly accepted subjective similarity measure. Results can vary depending on the similarity measures used.

Nonetheless, seemingly different similarity measures may be equivalent after some transformation.

Suppose we have the following two-dimensional data set:

| Data Point | $A_1$ | $A_2$ |
|------------|------:|------:|
| $x_1$      | 1.5   | 1.7   |
| $x_2$      | 2.0   | 1.9   |
| $x_3$      | 1.6   | 1.8   |
| $x_4$      | 1.2   | 1.5   |

#### (a)

Consider the data as two-dimensional data points.

Given a new data point, $x = (1.4, 1.6)$, as a query, rank the database points based on similarity with the query using:

- Euclidean distance
- Manhattan distance
- Supremum distance
- Cosine similarity

#### (b)

Normalize the data set to make the norm of each data point equal to 1.

Use Euclidean distance on the transformed data to rank the data points.


###Answer:

(a) Ranking the Database Points

1. Euclidean Distance ($L_2$ norm)


**Formula:**
$$
d(x, y) = \sqrt{(x_1 - y_1)^2 + (x_2 - y_2)^2}
$$

$$
x_1: \sqrt{(1.5-1.4)^2 + (1.7-1.6)^2} = \sqrt{0.02} \approx \mathbf{0.141}

## Problem 6

Compute the cosine measure using the raw frequencies between the following two sentences:

(a) “The sly fox jumped over the lazy dog.”

(b) “The dog jumped at the intruder.”

##Answer






all unique words are:

the, sly, fox, jumped, over, lazy, dog, at, intruder



Frequency Vectors is

| Word       | Sentence A ($x$) | Sentence B ($y$) |
|------------|------------------|------------------|
| the        | 2                | 2                |
| sly        | 1                | 0                |
| fox        | 1                | 0                |
| jumped     | 1                | 1                |
| over       | 1                | 0                |
| lazy       | 1                | 0                |
| dog        | 1                | 1                |
| at         | 0                | 1                |
| intruder   | 0                | 1                |

Vector $x$:  
$(2, 1, 1, 1, 1, 1, 1, 0, 0)$  

Vector $y$:  
$(2, 0, 0, 1, 0, 0, 1, 1, 1)$  





$$
sim(x, y) = \frac{x \cdot y}{\|x\| \|y\|}
$$

1. ($x \cdot y$) =

$$
(2 \times 2) + (1 \times 0) + (1 \times 0) + (1 \times 1) + (1 \times 0) + (1 \times 0) + (1 \times 1) + (0 \times 1) + (0 \times 1)
$$

$$
= 4 + 0 + 0 + 1 + 0 + 0 + 1 + 0 + 0 = \mathbf{6}
$$



2. Magnitude of $x$

$$
\|x\| = \sqrt{2^2 + 1^2 + 1^2 + 1^2 + 1^2 + 1^2 + 1^2}
$$

$$
= \sqrt{4 + 1 + 1 + 1 + 1 + 1 + 1} = \sqrt{10} \approx \mathbf{3.162}
$$



3. Magnitude of $y$

$$
\|y\| = \sqrt{2^2 + 1^2 + 1^2 + 1^2 + 1^2}
$$

$$
= \sqrt{4 + 1 + 1 + 1 + 1} = \sqrt{8} \approx \mathbf{2.828}
$$



so,

$$
sim(x, y) = \frac{6}{3.162 \times 2.828}
$$

$$
= \frac{6}{8.942} \approx \mathbf{0.671}
$$
