```python
import torch
import torch.nn as nn #component for building neaural network
import torch.optim as optim #tools for training networks
```

```python
#distance in miles
distance = torch.tensor([[1.0],[2.0],[3.0],[4.0]], dtype=torch.float32)

#[1.0],[2.0],[3.0],[4.0] these are collection of indivijual data points, this
#colllection is called batch, inner set [1.0] is called one sample in a batch
#here each sample has only one feature which is distance

#dtype is telling python its 32- bit floating point number to store

#Delivery times in minutes
times = torch.tensor([[6.96], [12.11],[16.77],[22.21]], dtype=torch.float32)
```

```python
#Define the model
model = nn.Sequential(nn.Linear(1,1))

#here we are using single neuron so one layer which is linear
#(1,1) means its taking one input and producing one output.
#this neuron will learn best weights and best bias

#Sequential is the pytorch streamlined version, its container that passes data through layers in order
```

```python
#Define the loss function
loss_function = nn.MSELoss()

#MSE = mean squared error loss, it measures how wrong and how right the predictions are
```

```python
#Define the optimizer
optimizer = optim.SGD(model.parameters(), lr=0.01)

#SGD is stochastic gradient descent is the algo that figure out
#which direction to adjust weights and bias to reduce the error

#model.parameters is how to access weights and bias values

#lr = learning rate, smaller value means adjusted the w and b parameters by smaller amount,
# larger lr means bigger adjustment
```

```python
#Training loop

for epoch in range(500): #500 times it is tring to fit the line
  #each full passs to the traning data is called epoch
  #Reset the optimizer
  optimizer.zero_grad() #optimize all the tarning value from the
  #previous round
  #1.make predictions
  outputs = model(distance) #output = model(input)
  #tells the model use distance as inputs
  #2. calculate the loss- how bas was the guess?
  loss = loss_function(outputs, times)
  #compare the predictions with the real time
  #calculate adjustments
  #adjust the internal parameters to improve the predictions
  loss.backward() #backpropagation
  #update the model
  optimizer.step()
```

```python
#inference

with torch.no_grad(): #tells pytorch that not traning anymore just inference
    test_distance = torch.tensor([[25.0]], dtype=torch.float32)
    predicted_time = model(test_distance)
    print(f"Predicted time for 25 miles: {predicted_time.item():.1f}minutes")
```
