[LP 1] Deal with the missing values in the credit analysis dataset. Explain what you did with the following columns:

1. Column A1 (at index 0)
2. Column A2 (at index 1)
3. Column A4 (at index 3)
4. Column A5 (at index 4)
5. Column A6 (at index 5)
6. Column A7 (at index 6)
7. Column A14 (at index 13)



```python
## Your code for LP 1 goes here.

column_names = [
    "A1", "A2", "A3", "A4", "A5", "A6", "A7", "A8",
    "A9", "A10", "A11", "A12", "A13", "A14", "A15", "A16"
]


df = pd.read_csv('/content/crx (1).data', names=column_names, na_values="?")

categorical_cols = ["A1", "A4", "A5", "A6", "A7"]
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])


numerical_cols = ["A2", "A14"]
for col in numerical_cols:

    df[col] = pd.to_numeric(df[col], errors='coerce')
    df[col] = df[col].fillna(df[col].mean())

print("Missing values after imputation:")
print(df[["A1", "A2", "A4", "A5", "A6", "A7", "A14"]].isnull().sum())
```


[LP 2] Use encoders and convert the categorical variables to numerical values: Columns A1, A4, A5, A6, A7, A9, A10, A12, A13


```python
## Your code for LP 2 goes here.

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

cols_to_encode = ["A1", "A4", "A5", "A6", "A7", "A9", "A10", "A12", "A13"]

for col in cols_to_encode:

    if df[col].isnull().any():
        df[col] = df[col].fillna(df[col].mode()[0])


    df[col] = le.fit_transform(df[col].astype(str))

print("Categorical columns after encoding:")
print(df[cols_to_encode].head())

print(df[cols_to_encode].dtypes)
```

[HP 1] Describe your choices to preprocess these columns and explain why they are appropriate for the column.

For the High Pass (HP), the  data cleaning to ensuring the dataset is mathematically optimized for machine learning algo.

1. Categorical Imputation: The Mode (A1, A4, A5, A6, A7)
Choice: Imputing missing values using the most frequent value (Mode).
Categorical data like occupation (A6) or marital status (A4) does not have a mathematical average (Mean). Using the Mode ensures that we fill gaps with a value that already exists in the real-world distribution of the dataset. The mode is the "safest bet" in a classification context because it assumes the missing data follows the trend of the majority, preventing the creation of "outlier" categories that could confuse a model.

2. The Mean (A2, A14) Impute missing values using the Arithmetic Mean.For continuous variables like age (A2), the mean provides a balanced representative value of the entire column.
In the Credit Approval dataset, the numerical ranges for age and zip code (A14) are relatively stable. Using the mean minimizes the variance introduced by missing data, which helps maintain the statistical properties required for algorithms.

3. Label Encoding (A1, A4, A5, A6, A7, A9, A10, A12, A13) is transforming strings into integers via Label Encoding. Machine learning models (like those in scikit-learn) perform calculations on matrices. They cannot perform "multiplication" or "distance calculation" on strings like "v" or "h". Encoding converts these into a computer-readable format.
Since many of these columns are binary, Label Encoding is highly memory-efficient, converting them into simple 0s and 1s without expanding the number of columns in the dataset.

4.The goal of this dataset is to predict credit approval. The positive class (approval) is typically represented by 1 and the negative class (rejection) by 0. This is necessary for calculating performance metrics like precision, recall, and the F1-score.



[LP 3] Use an appropriate scaler to scale the numerical values to a suitable range of values. Briefly explain the process you used.

```python
## Your code for LP 3 goes here.

from sklearn.preprocessing import StandardScaler

numerical_cols = ["A2", "A3", "A8", "A11", "A14", "A15"]
scaler = StandardScaler()

df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

print("numerical columns after Standard Scaling:")
print(df[numerical_cols].head())

print("scaling verification (Mean)")
print(df[numerical_cols].mean())
print("scaling verification (Std Dev)")
print(df[numerical_cols].std())
```

For [LP 3], I used the StandardScaler from scikit-learn to normalize the numerical features. This process ensures that all numerical variables contribute equally to the model, preventing features with large magnitudes from dominating those with smaller ranges.
